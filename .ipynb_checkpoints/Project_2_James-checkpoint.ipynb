#### There is an unwanted element '-' in our column. We can simply use loc to locate its column and row, and assign a new value to it. 

````python
drugs.loc[drugs['cocaine-frequency']=='-', 'cocaine-frequency']
````

#### But for practice purpose, let's use a mask instead

````python
cocaine_frequency_mask = drugs['cocaine-frequency'] == '-'
cocaine_frequency_mask
````

````python
drugs.loc[cocaine_frequency_mask, 'cocaine-frequency'] = 0
````

````python
drugs['cocaine-frequency'] = drugs['cocaine-frequency'].map(lambda x: float(x))
````

### Works great. Let's make a function to do this for the other cols.

````python
def mask(db_name, col_name, search_for, replace_with):
    new_mask = db_name[col_name] == search_for
    db_name.loc[new_mask, col_name] = replace_with
    db_name[col_name] = db_name[col_name].map(lambda x: float(x))
````

````python
mask(drugs, 'heroin-frequency', '-', 0)
mask(drugs, 'inhalant-frequency', '-', 0)
mask(drugs, 'oxycontin-frequency', '-', 0)
mask(drugs, 'meth-frequency', '-', 0)
````

### Check to see if any columns are not int or float

### James' Answer

Let's engineer a new feature.

To measure dependency on a drug, i will create a measure called "reliance". Calculated as 'frequency/usage'. 

If frequency is low and usage is high, it is similar to a party drug. Many people have used it before, but there is low reliance on the drug.

### Forming my null and alternative hypothesis

From eyeballing the data, it seems that older users have a higher reliance on pain reliver.

Null hypothesis: The difference in 'reliance on pain reliever' between those aged >=40 and those aged < 40 is zero.

Alternative hypothesis: The mean in 'reliance on pain reliver' between those aged >=40 and < 40 is not zero.

 T-test with alpha of 0.05. Some rules to break:  The distribution in 'reliance' might not be normally distributed, but we will go ahead with t-test anyway. 
 
Why T-Test?
 
*A t-test is an analysis of two populations means through the use of statistical examination; a t-test with two samples is commonly used with small sample sizes, testing the difference between the samples when the variances of two normal distributions are not known.*


#### Create pain_df which contains the info i need to work with. Split into 2 groups, age >=40 and age < 40

````python
#pain_df = drugs.loc[:, ['age','n','pain-releiver-use','pain-releiver-frequency']]
pain_under40 = drugs.loc[0:13, ['age','n','pain-releiver-use','pain-releiver-frequency']]

pain_under40
pain_under40['reliance'] = pain_under40['pain-releiver-frequency'] / pain_under40['pain-releiver-use']
pain_under40['weight'] = pain_under40['n'] / pain_under40['n'].sum()
pain_under40['weighted_reliance'] = pain_under40.reliance * pain_under40.weight
under40_mean = pain_under40.weighted_reliance.sum()
print 'the mean reliance for our under 40 group is: ' + str(under40_mean)

pain_under40['squared_dist_from_mean'] = (pain_under40.reliance - pain_under40.weighted_reliance.sum() ) ** 2

under40_variance = pain_under40['squared_dist_from_mean'].sum() / len(pain_under40.index)
print 'the variance in reliance for our under 40 group is ' + str(under40_variance) 
print 'the std dev is ' + str(np.sqrt(under40_variance))
print 'the n for our under 40 group is ' + str(len(pain_under40.index))
````

````python

pain_over40 = drugs.loc[14:, ['age','n','pain-releiver-use','pain-releiver-frequency'] ]
pain_over40

pain_over40
pain_over40['reliance'] = pain_over40['pain-releiver-frequency'] / pain_over40['pain-releiver-use']
pain_over40['weight'] = pain_over40['n'] / pain_over40['n'].sum()
pain_over40['weighted_reliance'] = pain_over40.reliance * pain_over40.weight
over40_mean = pain_over40.weighted_reliance.sum()
print 'the mean reliance for our over 40 group is: ' + str(over40_mean)

pain_over40['squared_dist_from_mean'] = (pain_over40.reliance - pain_over40.weighted_reliance.sum() ) ** 2

over40_variance = pain_over40['squared_dist_from_mean'].sum() / len(pain_over40.index)
print 'the variance in reliance for our over 40 group is ' + str(over40_variance) 
print 'the std dev is ' + str(np.sqrt(over40_variance))
print 'the n for our over 40 group is ' + str(len(pain_over40.index))
````

#### Solve for critical t value

````python
#numerator is (under40_mean - over40_mean)
#denominator is np.sqrt((under40_variance/len(pain_under40.index))  + (over40_variance/len(pain_over40.index)))
t_statistic = (over40_mean - under40_mean) / np.sqrt((under40_variance/len(pain_under40.index))  + (over40_variance/len(pain_over40.index)))

critical_value = 1.753 #taken from t-table

print 'our t statistic is ' + str(t_statistic) 
print 'degrees of freedom is ' + str(len(pain_over40.index) + len(pain_under40.index) - 2)
print 'crtical value is ' + str(critical_value)

# if t > critical_value, the signals are significance.
#In this case, t statistic is less than critical value. Hence, accept null.
````

#### t statistic < critical value. Hence, do not reject the null hypothesis.

#### Spearman correlation is used when 2 variables are correlated, but not linearly correlated. 

The variables in the spearman correlation formula are the "rank" of our variables, (i.e integers ranging from 1 to n).  Spearman correlation is a value of +1 or -1

````python

stats.spearmanr(sat[['Math','Verbal']])
print 'The Spearman correlation of Math and Verbal is ' + str(stats.spearmanr(sat[['Math','Verbal']])[0])

# Let's compare to pearson's correlation.

stats.pearsonr(sat.Math, sat.Verbal)
print 'The pearsons correlation of Math and Verbal is ' +  str(stats.pearsonr(sat.Math, sat.Verbal)[0])

#The difference is because spearmans correlation is less affected by outliers. 

````
